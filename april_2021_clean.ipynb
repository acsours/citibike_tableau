```python
import pandas as pd
```

```python
file="resources/JC-202104-citibike-tripdata.csv"

df=pd.read_csv(file)
df.head()
```

```python
df.count()
```

```python
df.dtypes
```

```python
df.isnull().any().any()
```

```python
df=df.drop_duplicates()
df.count()
```

```python
df.head()
```

```python
# if start and end are the same station, filter
filter_df=df.loc[:, 'start station id']==df['end station id']
repeat_df=df.loc[filter_df, :]
repeat_df.count()
```

```python
repeat_df.sort_values(['tripduration'], ascending=True).head(20)
```

```python
# filter where tripduration < 120 seconds
filter_time_df=repeat_df.loc[:, 'tripduration']<120
short_df=repeat_df.loc[filter_time_df, :]
short_df.count()
```

```python
short_df.sort_values(['tripduration'])
```

```python
filter_1=df.loc[:, 'start station id']!=df['end station id']
filter_2=df.loc[:, 'tripduration']>120
new_df=df.loc[filter_1 & filter_2, :]
new_df.count()
```

```python
# write to csv
new_df.to_csv('resources/april_2021_clean.csv', index=False, header=True)
```
